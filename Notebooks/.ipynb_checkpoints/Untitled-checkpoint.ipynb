```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_restaurants = pd.read_csv('../data/df_restaurants.csv')
```

```python
df_restaurants
```

```python
rating_mean = df_restaurants.groupby('City').Rating.mean()
sort_rating_mean = rating_mean.sort_values(ascending=False)
```

```python
sort_rating_mean = pd.DataFrame(sort_rating_mean)
```

```python
sort_rating_mean
```

```python
top_rating_mean = sort_rating_mean.head()
```

```python
top_rating_mean
```

```python
sns.lineplot(x=top_rating_mean.index, y = top_rating_mean.Rating, hue=top_rating_mean.index)
```

```python
sns.boxplot(x="Price Range", y="Rating", data=df_restaurants)
```

```python
sns.scatterplot(x=df_restaurants.index, y = df_restaurants.Rating, hue=top_rating_mean.index)
```

```python
df_pais = df_restaurants[df_restaurants.City=='Athens']
```

```python
df_pais
```

```python
sns.histplot(x=df_pais['Price Range'], hue = df_pais['Price Range'])
```
